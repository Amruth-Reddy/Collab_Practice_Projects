```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
path = "/content/diabetes_prediction_dataset.csv"
df = pd.read_csv(path)
```

```python
print(" First 5 Rows of the dataset:")
print(df.head())
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
print("\nClass distribution (target variable):")
print(df['diabetes'].value_counts())
```

```python
df.hist(figsize=(12, 10), bins=30)
plt.suptitle("Feature Distributions", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot of Features")
plt.show()
```

```python
df_numeric = df.select_dtypes(include=[np.number])  # Keep only numeric columns
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue="diabetes")
plt.show()
```
